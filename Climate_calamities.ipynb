# This dataset provides information about the Global Climate Risk Index (CRI) and associated economic losses for various countries around the world. The Global Climate Risk Index measures a countryâ€™s vulnerability and resilience to extreme climate related events such as floods, heatwaves, storms, droughts and wildfires. The CRI takes into account fatalities caused by these events as well as economic losses in purchasing power parity US dollars.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('climate-risk-index-1.csv')
```

```python
df.head()
```

```python
#REMOVING UNNECESSARY COLUMNS
df.drop(['the_geom','the_geom_webmercator'], axis=1, inplace=True)
```

```python
#RESETING INDEX
df.set_index(['cartodb_id'], inplace=True)
```

# Higher the rank, Higher the risk

```python
high_risk=df.nsmallest(n=10, columns=['cri_rank'])
high_risk
```

```python
plt.figure(figsize=(15,7))
plt.title('10 Highest Effected Countries')
plt.bar(high_risk['country'], high_risk['cri_rank'], color='red')
plt.xlabel('country')
plt.ylabel('cri_rank')
plt.show()
```

```python
#Fatalities in a year
df["fatalities_total"].sum()
```

```python
#top 10 countries with highest GDP loss
top_loser_gdp=df.nlargest(n=10, columns=['losses_per_gdp__total'])
top_loser_gdp
```

```python
plt.figure(figsize=(15,7))
plt.title('10 Countries with highest GDP Loss')
plt.bar(top_loser_gdp['country'], top_loser_gdp['losses_per_gdp__total'], color='red')
plt.xlabel('country')
plt.ylabel('total loss per GDP')
plt.show()
```

```python
#TOTAL GDP LOSS IN PPP TERMS
df['losses_usdm_ppp_total'].sum()
```

```python
#TOP GDP LOSERS IN PPP TERMS
top_ppp_loser=df.nlargest(n=10, columns=['losses_usdm_ppp_total'])
top_ppp_loser
```

```python
plt.figure(figsize=(15,7))
plt.title('10 Countries with highest GDP Loss in PPP')
plt.bar(top_ppp_loser['country'], top_ppp_loser['losses_usdm_ppp_total'], color='red')
plt.xlabel('country')
plt.ylabel('total loss per GDP in PPP')
plt.show()
```
